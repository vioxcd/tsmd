#### Simple exploration

##### longest tags in a session

```python
def list_len(tag):
    tag = eval(tag)
    return len(tag)

df.tag.apply(list_len).max()
```

##### tags with most urls

```python
from collections import defaultdict

counter = defaultdict(int)

def tag_counter(tag):
    tag = eval(tag)
    for t in tag:
        global counter
        counter[t] += 1
    return tag

df.tag.apply(tag_counter)
counter
```

```python
unique_urls_by_tags = sorted(counter.items(), key=lambda item: item[1], reverse=True)
unique_urls_by_tags[:10]
```
