##### Tags with longest span of time

```python
# nan, mail, pm, dm, _startup, gsp, aom, winClose, browserExit, prom, tts
```

```python
df_with_tags = (df
                .explode('tag')
                .reset_index(drop=True)
                .dropna(subset=['tag'])
               )
```

```python
unique_tags = (df_with_tags
              ['tag']
              .unique()
              )
unique_tags
```

```python
time_diff_per_tags = []
for tag in unique_tags:
    aggs = (df_with_tags
            .query(f"tag == '{tag}'")
            ['date']
            .agg([min, max])
           )
    time_diff_per_tags.append({
        "tag": tag,
        "time_diff": aggs.max() - aggs.min()
    })
```

```python
time_diff_per_tags_df = pd.DataFrame(time_diff_per_tags)
time_diff_per_tags_df.head()
```

```python
(time_diff_per_tags_df
 .sort_values('time_diff', ascending=False)
 .head(20)
)
```

##### longest tags in a session

```python
def list_len(tag):
    tag = eval(tag)
    return len(tag)

df.tag.apply(list_len).max()
```

##### tags with most urls

```python
from collections import defaultdict

counter = defaultdict(int)

def tag_counter(tag):
    tag = eval(tag)
    for t in tag:
        global counter
        counter[t] += 1
    return tag

df.tag.apply(tag_counter)
counter
```

```python
unique_urls_by_tags = sorted(counter.items(), key=lambda item: item[1], reverse=True)
unique_urls_by_tags[:10]
```
